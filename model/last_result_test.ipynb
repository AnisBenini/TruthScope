```python
import json
import matplotlib.pyplot as plt

# === 1. Charger les donnes ===
with open(r"./metrics_log.json", "r") as f:
    logs = json.load(f)

# === 2. Extraire les mtriques ===
epochs = [entry["epoch"] for entry in logs]
validation_loss = [entry["eval_loss"] for entry in logs]
accuracy = [entry["eval_accuracy"] for entry in logs]
f1 = [entry["eval_f1"] for entry in logs]
precision = [entry["eval_precision"] for entry in logs]
recall = [entry["eval_recall"] for entry in logs]

plt.style.use("seaborn-v0_8-darkgrid")

# === 3. Courbe : Validation Loss ===
plt.figure(figsize=(8, 5))
plt.plot(epochs, validation_loss, marker='o', label='Validation Loss')
plt.title("volution de la perte de validation")
plt.xlabel('poque')
plt.ylabel('Loss')
plt.legend()
plt.tight_layout()
plt.show()

# === 4. Courbe : Accuracy ===
plt.figure(figsize=(8, 5))
plt.plot(epochs, accuracy, marker='o', color='green', label='Accuracy')
plt.title("volution de la prcision (Accuracy)")
plt.xlabel('poque')
plt.ylabel('Accuracy')
plt.legend()
plt.tight_layout()
plt.show()

# === 5. Courbe : F1 Score ===
plt.figure(figsize=(8, 5))
plt.plot(epochs, f1, marker='o', label='F1 Score')
plt.title("volution du F1 Score")
plt.xlabel('poque')
plt.ylabel('Score')
plt.legend()
plt.tight_layout()
plt.show()

# === 6. Courbe : Precision ===
plt.figure(figsize=(8, 5))
plt.plot(epochs, precision, marker='s', label='Precision', color='orange')
plt.title("volution de la prcision")
plt.xlabel('poque')
plt.ylabel('Score')
plt.legend()
plt.tight_layout()
plt.show()

# === 7. Courbe : Recall ===
plt.figure(figsize=(8, 5))
plt.plot(epochs, recall, marker='^', label='Recall', color='purple')
plt.title("volution du rappel (Recall)")
plt.xlabel('poque')
plt.ylabel('Score')
plt.legend()
plt.tight_layout()
plt.show()
```
